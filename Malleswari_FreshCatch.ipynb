```python
data.drop(columns=['Lineitem taxable','Lineitem requires shipping','Currency',
                   'Vendor','Country','Zip','Accepts SMS Marketing'],inplace=True)
```

```python
data.drop(columns=['Created at','Payment Reference','Risk Level'],inplace=True)
```

```python
data['Tags_x']=data['Tags_x'].replace('android, Login with Phone Number',"Others")
data['Tags_x']=data['Tags_x'].replace('android, iphone',"Others")
data['Tags_x']=data['Tags_x'].replace('iphone, newsletter',"Others")
data['Tags_x']=data['Tags_x'].replace('android, newsletter',"Others")
data['Tags_x']=data['Tags_x'].replace('iphone, Login with Phone Number',"Others")
data['Tags_x']=data['Tags_x'].replace('Outlet',"Others")
data['Tags_x']=data['Tags_x'].replace('password page, prospec',"Others")
data['Tags_x']=data['Tags_x'].replace('Phone Number',"Others")
```

```python
data['Shipping Method']=data['Shipping Method'].replace('Standard','Standard Shipping')
```

```python
round(data.isnull().sum()/len(data)*100,2)
```

```python
data['Discount Amount']=data['Discount Amount'].fillna(0)
```

```python
data['Accepts Marketing'].unique()
```

```python
data['Accepts Marketing']=data['Accepts Marketing'].fillna('yes')
```

```python
data['Shipping'].unique()
```

```python
data['Taxes'].unique()
```

```python
data['Taxes']=data['Taxes'].fillna(0)
```

```python
data['Total']=data['Total'].fillna(data['Lineitem quantity']*data['Lineitem price'])
```

```python
data['Discount Amount']=data['Discount Amount'].fillna(0)
```

```python
data['Shipping Method'].unique()
```

```python
data['Shipping Method']=data['Shipping Method'].fillna("others")
```

```python
data['Refunded Amount']=data['Refunded Amount'].fillna(0)
```

```python
data['Outstanding Balance']=data['Outstanding Balance'].fillna(0)
```

```python
data['Source']=data['Source'].replace('android','others')
data['Source']=data['Source'].replace('1830279','others')
data['Source']=data['Source'].fillna('others')
```
